```python
import pandas as pd

#loading the dataset from mydrive

data = pd.read_csv('/content/drive/MyDrive/creditcard_project/Fresh_creditcard.csv')

data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#removing the duplicates

#before removing the duplicates
print('Initially:', data.shape)

#drop dulpicates
data = data.drop_duplicates()

#after removing the duplicates
print('After removing duplicates:', data.shape)
```

```python
#counting fraud vs non fraud counts
fraud_counts = data['Class'].value_counts()
print(fraud_counts)
```

```python
#Representing with a bar chart
import matplotlib.pyplot as plt

fraud_counts.plot(kind='bar', color=['pink', 'purple'])
plt.title('FRAUD VS NON-FRAUD TRANSACTIONS')
plt.xlabel('TRANSACTION TYPE (0= NON-FRAUD, 1= FRAUD)')
plt.ylabel('Count')
plt.show()
```

```python
#using pie chart to display this as well
plt.pie(fraud_counts, labels=['NON-FRAUD(0)', 'FRAUD(1)'], startangle=120, colors=['orange', 'purple'], autopct='%1.2f%%')
plt.title('FRAUD VS NON-FRAUD TRANSACTIONS')
plt.show()
```
